DATA PREPROCESSING

```python
len(car['car_ID'])
```

```python
car.drop(['car_ID'],axis=1,inplace=True)
```

```python
car['symboling'].unique()
```

```python
print(len(car['CarName'].unique()))
car['CarName'].unique()
```

```python
car['Company'] = car.CarName.str.split(expand=True)[0]
```

```python
del car['CarName']

#company
car['Company'].unique()
```

```python
car.replace(['maxda','nisan','porsche','toyota', 'volkswagen', 'vw'],['mazda','Nissan','porsche','toyata','volkswagen','volkswagen'])
car['Company'].unique()
```

```python
#label encoder
```

```python
label_encoder = preprocessing.LabelEncoder()

car['symboling']=label_encoder.fit_transform(car['symboling'])
car['fueltype']=label_encoder.fit_transform(car['fueltype'])
car['aspiration']=label_encoder.fit_transform(car['aspiration'])
car['doornumber']=label_encoder.fit_transform(car['doornumber'])
car['carbody']=label_encoder.fit_transform(car['carbody'])
car['drivewheel']=label_encoder.fit_transform(car['drivewheel'])
car['enginlocation']=label_encoder.fit_transform(car['enginelocation'])
car['Company']=label_encoder.fit_transform(car['Company'])

```

```python
print(len(car.columns))
car.columns
```

```python

#feature selection
X=car[['symboling','fueltype','doornumber','carheight', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg', 'price']]
Y=car[['price']]
X.head()

```

DATA SPLITTING

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
print('x_train:' ,x_train.shape)
print('x_test: ',x_test.shape)
print('y_train:',y_train.shape)
print('y_test:',y_test.shape)
```

MODEL SELECTION AND IMPLEMENTATION

```python
clf = LinearRegression()
```

MODEL TRAINING

```python
clf.fit(x_train,y_train)
```

MODEL EVALUATION

```python
clf.score(x_test,y_test)
```

```python
predict = clf.predict(x_test)
predict
```
